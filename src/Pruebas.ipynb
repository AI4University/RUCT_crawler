```python
def obtener_modulos(url_base, codigoin):
    url = url_base.replace('codigoin', codigoin)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    modulos = []

    # Encontrar la tabla con id 'modulo'
    tabla_modulos = soup.find('table', {'id': 'modulo'})
    if tabla_modulos:
        # Iterar sobre las filas del cuerpo de la tabla
        for fila in tabla_modulos.find('tbody').find_all('tr'):
            # Extraer la primera celda (que contiene la ID del módulo)
            celdas = fila.find_all('td')
            if celdas:
                modulo_id = celdas[0].text.strip()
                modulos.append(modulo_id)
    else:
        print(f"No se encontró la tabla de módulos para el código {codigoin}")

    return modulos
```

```python
def obtener_materias_por_modulos(url_base, codigoin, ids_modulos):
    diccionario_materias = {}
    nombres_materias = []
    
    with ThreadPoolExecutor(max_workers=16) as executor:
        futures = []
        for codModulo in ids_modulos:
            url = url_base.replace('codigoModulo', str(codModulo)).replace('codigoin', codigoin)
            futures.append(executor.submit(obtener_materias_por_modulo, url))
        
        for future in as_completed(futures):
            try:
                ids_materias, nombres = future.result()
                diccionario_materias[codModulo] = ids_materias
                nombres_materias.extend(nombres)
            except Exception as e:
                print(f"Error obtaining materias for module {codModulo}: {e}")

    return diccionario_materias, nombres_materias
```

```python
def obtener_datos_asignatura(url_base, codModulo, codMateria, codigoin):
    url = url_base.replace('codigoModulo', str(codModulo)).replace('codigoMateria', str(codMateria)).replace('codigoin', codigoin)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    asignaturas = [] # hacer fuera para iteraciones consecutivas
    
    # Encontrar la tabla con id 'asignatura'
    tabla_asignaturas = soup.find('table', {'id': 'asignatura'})
    if tabla_asignaturas:
        # Iterar sobre las filas del cuerpo de la tabla
        for fila in tabla_asignaturas.find('tbody').find_all('tr'):
            # Extraer la segunda celda (que contiene el nombre de la asignatura)
            celdas = fila.find_all('td')
            if celdas:
                asignatura_nombre = celdas[1].text.strip()
                asignaturas.append(asignatura_nombre)
    
    return asignaturas
```

```python
import re

def limpiar_texto(texto):
    # Eliminar saltos de línea y caracteres especiales
    texto_limpio = re.sub(r'\s+', ' ', texto)  # Reemplazar cualquier espacio en blanco (incluye \n, \t) por un espacio
    texto_limpio = re.sub(r'[^a-zA-Z0-9áéíóúÁÉÍÓÚñÑüÜ\s]', '', texto_limpio)  # Eliminar caracteres especiales, excepto letras y números
    return texto_limpio.strip()

def obtener_contenidos(url_base, codModulo, codMateria, codigoin):

    url = url_base.replace('codigoModulo', str(codModulo)).replace('codigoMateria', str(codMateria)).replace('codigoin', codigoin)
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the "Contenidos de la materia" section
    contents_section = soup.find('fieldset', id='datosMateria_contenidos').find('div', class_='texto-editor')
    
    # Combine all the contents into a single string, joined by a delimiter like newlines or spaces
    combined_contents = ' '.join([limpiar_texto(content.text) for content in contents_section if content.text.strip()])
    
    return combined_contents

url_base = 'https://www.educacion.gob.es/ruct/solicitud/datosMateria!contenidos?tipo=C&actual=menu.solicitud.planificacion.materias.contenidos&codMateria=codigoMateria&codModulo=codigoModulo&cod=codigoin'

prueba1 = obtener_contenidos(url_base, "1", "1", "25039172018113001")
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed

def get_contenidos(url_base, id):
    url_base_modulos = 'https://www.educacion.gob.es/ruct/solicitud/modulos?actual=menu.solicitud.planificacion.modulos&cod=codigoin'
    url_base_materias = 'https://www.educacion.gob.es/ruct/solicitud/datosModulo?codModulo=codigoModulo&actual=menu.solicitud.planificacion.materias&cod=codigoin'
    ids_modulos = obtener_modulos(url_base_modulos, id)
    diccionario_materias, _ = obtener_materias_por_modulos(url_base_materias, id, ids_modulos)

    contenidos_lista = []
    with ThreadPoolExecutor(max_workers=16) as executor:  # Usar 16 hilos para aprovechar el rendimiento de la M2
        futures = []
        for codModulo, ids_materias in diccionario_materias.items():
            for codMateria in ids_materias:
                futures.append(executor.submit(obtener_contenidos, url_base, codModulo, codMateria, id))
        
        for future in as_completed(futures):
            try:
                contenidos = future.result()
                contenidos_lista.append(contenidos)
            except Exception as e:
                print(f"Error obtaining contents: {e}")

    return contenidos_lista
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir la columna "calendario" a tipo datetime para extraer el año
data['calendario'] = pd.to_datetime(data['calendario'], format='%Y').dt.year

# Filtrar los datos para incluir solo los años desde 2000 en adelante
data = data[data['calendario'] >= 2000].copy()  # Usamos .copy() para evitar SettingWithCopyWarning

# Definir función para verificar si la información es completa o incompleta
def verificar_informacion(columna):
    return data[columna].apply(lambda x: not (x in ["No encontrado", "[]", ""] or len(eval(x)) == 0 if isinstance(x, str) else len(x) == 0))

# Crear columnas que indiquen si la información está completa o incompleta para cada categoría usando .loc[]
data.loc[:, 'Contenidos_completo'] = verificar_informacion('Contenidos')
data.loc[:, 'Modulo_completo'] = verificar_informacion('Modulo')
data.loc[:, 'Materias_completo'] = verificar_informacion('Materias')
data.loc[:, 'Asignaturas_completo'] = verificar_informacion('Asignaturas')

# Listado de las columnas que queremos analizar
columnas = ['Contenidos_completo', 'Modulo_completo', 'Materias_completo', 'Asignaturas_completo']
titulos = ['Contenidos', 'Modulo', 'Materias', 'Asignaturas']

# Crear una figura con 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Iterar sobre las columnas y los ejes para crear los histogramas apilados
for i, (columna, titulo) in enumerate(zip(columnas, titulos)):
    # Contar el número de grados con información completa e incompleta por año
    completos_por_anio = data.groupby('calendario')[columna].sum()
    incompletos_por_anio = data.groupby('calendario').size() - completos_por_anio

    # Determinar la posición del subplot
    ax = axes[i//2, i%2]
    
    # Crear el histograma apilado
    ax.bar(completos_por_anio.index, completos_por_anio, label=f'{titulo} Completos')
    ax.bar(completos_por_anio.index, incompletos_por_anio, bottom=completos_por_anio, label=f'{titulo} Incompletos')
    
    # Configurar el título y las etiquetas
    ax.set_title(f'Histogram of {titulo} with complete and incomplete information by year (2000 onwards)')
    ax.set_xlabel('Year')
    ax.set_ylabel('Number of Degrees')
    ax.legend()
    ax.grid(True)

# Ajustar el layout para que los subplots no se sobrepongan
plt.tight_layout()

# Mostrar la figura con los subplots
plt.show()
```
